## Here is the dataset with "all" entries for restaurants in Europe

We are going to use this Dataset for a Dashboard where we will try to gather some insights from the restaurant scene in Europe

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('tripadvisor_european_restaurants copy.csv')
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
#See the shape
df.shape
```

```python
#First thing I am going to do is drop the columns that I do not want from the beginning
df = df.drop(['address', 'original_location' , 'claimed', 'awards', 'popularity_detailed', 'price_range',
              'meals', 'special_diets', 'features', 'original_open_hours', 'open_days_per_week', 
              'open_hours_per_week', 'working_shifts_per_week', 'reviews_count_in_default_language'], axis=1)
df.shape
```

```python
#See the number of NaN we have in the different columns
df.isna().sum()
```

```python
#We check the percentages of NaN per column, we will drop keywords for the dashboard
nulls = pd.DataFrame(df.isna().sum()*100/len(df), columns=['percentage'])
nulls.sort_values('percentage', ascending = False)
```

```python
#drop the column keywords
df = df.drop(['keywords','atmosphere'], axis=1)
```

```python
#from the column top tags we will only stay with the firs value while also fillini NaNs
df['top_tags'] = df['top_tags'].str.replace("(,).*","")
df['top_tags'] = df['top_tags'].fillna('Not Specified')
```

```python
#Going to do the same as above but with cuisines
df['cuisines'] = df['cuisines'].str.replace("(,).*","").fillna('Not Specified')
```

```python
#Change the € symbol to ranges
df['price_level'] = df['price_level'].fillna('Not Specified')
df['price_level'] = df['price_level'].replace('€', '0-20')
df['price_level'] = df['price_level'].replace('€€-€€€', '20-45')
df['price_level'] = df['price_level'].replace('€€€€', '+45')
```

```python
df = df.drop_duplicates(subset=['restaurant_link'])
```

```python
df.to_csv('dashboard_rest_db.csv', index=False)
```
